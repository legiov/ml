```python
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 7)
disc_uni_dist = scipy.stats.randint(1, 7)
pmf = disc_uni_dist.pmf(x)
pmf
```

```python
cdf = disc_uni_dist.cdf(x)
cdf
```

```python
x = np.arange(1, 251)
disc_uni_dist = scipy.stats.randint(1, 251)
cdf = disc_uni_dist.cdf(x)
cdf[49]
```

```python
data = scipy.stats.bernoulli.rvs(size=500, p=0.6)
ax = sns.distplot(data, kde=False, color='dodgerblue')
ax.set(xlabel='Значение случайной величины', ylabel='Частота')
```

```python
unique, counts = np.unique(data, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
scipy.stats.binom.pmf(8, 10, 0.8) + scipy.stats.binom.pmf(9, 10, 0.8) + scipy.stats.binom.pmf(10, 10, 0.8)
```

```python
x = np.random.binomial(n=10, p=0.5, size=1000)
sns.histplot(x, kde=False)
plt.show()
```

```python
scipy.stats.distributions.poisson.pmf(5, 4.5)
```

```python
sns.histplot(np.random.poisson(lam=3, size=1000), kde=False)
plt.show()
```

```python
scipy.stats.distributions.poisson.pmf(16, 16)
```

```python
data = np.random.uniform(size=10000, low=10, high=40)
sns.displot(data, bins=30)
plt.show()
```

```python
data = scipy.stats.norm.rvs(size=10000, loc=4, scale=5)
sns.displot(data=data, bins=50)
```
