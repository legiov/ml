```python
import numpy as np
import pandas as pd

from sklearn import model_selection
from sklearn import ensemble
from sklearn import metrics
from sklearn import linear_model
from sklearn import tree
```

```python
df = pd.read_csv('data/weatherAUS.csv')
df
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
for col in df.columns:
    if df[col].isnull().sum()/df.shape[0]*100 > 40:
        df.drop(col, axis=1, inplace=True)
```

```python
df
```

```python
df['RainToday'].value_counts()
```

```python
df['RainToday'] = df['RainToday'].apply(lambda val: 1 if val == 'Yes' else 0)
df['RainTomorrow'] = df['RainTomorrow'].apply(lambda val: 1 if val == 'Yes' else 0)
```

```python
df['RainToday'].mean()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df = df.drop('Date', axis=1)
```

```python
df.groupby('Month')['RainToday'].agg('mean')
```

```python
cat = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'Month']
df = pd.get_dummies(df, columns=cat)
df.shape
```

```python
df.dropna(inplace=True)
X = df.drop('RainTomorrow', axis=1)
y = df['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=31)
```

```python
y_test.mean()
```

```python
np.random.seed(31)
mean_min_temp = []
for i in range(1000):
    rand_indexes = np.random.randint(1, X_train.shape[0], size=X_train.shape[0], )
    min_temp = X_train['MinTemp'].values
    mean_min_temp.append(np.mean(min_temp[rand_indexes]))
    
print(np.std(mean_min_temp))
```

```python
def fit_predict_and_print_metrics(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    print(metrics.roc_auc_score(y_train, y_train_pred))
    print(metrics.roc_auc_score(y_test, y_test_pred)) 
```

```python
lr = linear_model.LogisticRegression()
fit_predict_and_print_metrics(lr, X_train, X_test, y_train, y_test)
```

```python
params = {
    'max_leaf_nodes': list(range(2, 10)),
    'min_samples_split': [2, 3, 4],
    'max_depth': [5,7,9,11]
}
dt = tree.DecisionTreeClassifier(random_state=42)
gscv = model_selection.GridSearchCV(
    estimator=dt,
    param_grid=params,
    scoring='roc_auc',
    cv=3,
    verbose=3
)
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
model = tree.DecisionTreeClassifier(**gscv.best_params_, random_state=42)
fit_predict_and_print_metrics(model, X_train, X_test, y_train, y_test)
```

```python
model = ensemble.RandomForestClassifier(
    n_estimators=100,
    random_state=31
)

fit_predict_and_print_metrics(model, X_train, X_test, y_train, y_test)
```

```python
params = {'max_features': [ 4, 5, 6, 7], 'min_samples_leaf': [3, 5, 7, 9, 11], 'max_depth': [5, 10, 15]}
model = ensemble.RandomForestClassifier(
    n_estimators=100,
    random_state=31
)
gscv = model_selection.GridSearchCV(
    estimator=model,
    cv=3,
    param_grid= params,
    scoring='roc_auc'
)
gscv.fit(X_train, y_train)
```

```python
model = ensemble.RandomForestClassifier(**gscv.best_params_, n_estimators=100, random_state=31)

fit_predict_and_print_metrics(model, X_train, X_test, y_train, y_test)
```
